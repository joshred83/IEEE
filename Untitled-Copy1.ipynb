```python
def group_strong_corr(pd_corr_series, threshhold=0.95):
    # takes a column from a correlation matrix and reduces is it 
    # by rolling up all of the correlated columns into a list.
    # It returns a list of the columns which exceed the given
    #threshhold for correlation. 
    
    ans = pd_corr_series[(pd_corr_series > threshhold)]
    correlated_col_summary = ans.index.to_list() + [pd_corr_series.name]
    return correlated_col_summary

def group_strong_corr(pd_corr_series, threshhold=0.95):
    # takes a column from a correlation matrix and reduces is it 
    # by rolling up all of the correlated columns into a list.
    # It returns a list of the columns which exceed the given
    #threshhold for correlation. 
    
    ans = pd_corr_series[(pd_corr_series > threshhold)]
    correlated_col_summary = ans.index.to_list() + [pd_corr_series.name]
    return correlated_col_summary

def least_nulls(df, group_list):
    #takes a DataFrame and a list of columns in the DataFrame
    #returns the name of the column with the least null values
    return  df[group_list].isna().sum().sort_values().index[0]

def most_var(df, group_list):
    #takes a DataFrame and a list of columns in the DataFrame
    #returns the name of the column with the least null values
    return  df[group_list].var().sort_values(ascending=False).index[0]

def null_structure_encoder(df, group_list):
    # takes a dataframe, and a group of columns within the dataframe
    # identifies null values, and concatenates the results into a single column
    
    null_struct = df[group_list].isna().astype("int64")
    null_struct = null_struct[group_list].apply(lambda row: ''.join(row.values.astype(str)), axis=1)
    if null_struct.nunique() > 1:
        return null_struct

def consolidate_low_freq(pd_series, threshold=30):
    counts = pd_series.value_counts()
    low_counts = counts[counts <=threshold].index
    replace_map = {k:"other" for k in low_counts}
    return pd_series.replace(replace_map)

def remap_categorical_to_ordinal(col, target, consolidate=True, consolidate_threshhold=30):
    # takes a column of categorical variables, relates them to the target column and remaps them as
    # ordinal variables. Returns the remapped column and the mapper
    if consolidate:
        col = consolidate_low_freq(col, consolidate_threshhold)
    means = pd.DataFrame({'code':col, "isFraud":target}).groupby("code").mean().sort_values("isFraud")
    mapper = {code:val for code, val in zip(means.index, range(len(means.index)))}
    return col.replace(mapper), mapper
```

```python
correlated_col_summary = corr[corr.abs().max() > .95].agg(group_strong_corr, axis=1)

correlated_col_summary
```

```python
picks_from_correlations = sorted(list(
    {least_nulls(tx_train, group_list) for group_list in correlated_col_summary}.union(
    {most_var(tx_train, group_list) for group_list in correlated_col_summary} )
    
))
```

```python
X_train = tx_train[picks_from_correlations]
X_test = tx_test[picks_from_correlations]
```

```python
%%time
null_struct_train_cols =[]
null_struct_test_cols =[]
denom = len(correlated_col_summary)
for ix, group in enumerate(correlated_col_summary):
    print(f"{ix+1}/{denom}")
    null_struct = null_structure_encoder(tx_train, group_list=group)
    
    if null_struct is not None:
        null_struct_test = null_structure_encoder(tx_test, group_list=group)
        null_struct, mapper = remap_categorical_to_ordinal(null_struct, y_train, consolidate=True, consolidate_threshhold=30)
        null_struct_test = null_struct_test.replace(mapper)
        null_struct_train_cols.append( null_struct.rename(f"null_{ix}"))
        null_struct_test_cols.append( null_struct_test.rename(f"null_{ix}"))


    
```

```python
X_train = pd.concat([X_train]+null_struct_train_cols, axis=1)
X_test = pd.concat([X_test]+null_struct_test_cols, axis=1)
```
